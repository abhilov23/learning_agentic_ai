```python
# list comprehension is a concise way to create lists in Python.
#  It allows you to generate a new list by applying an expression to each item in an iterable, while optionally filtering items using a condition.
names = [
    "abhilov",
    "aditya",
    "jayesh",
    "aakash"
]

# basically we are checking if "abhi" is present in the name or not and if it is present then we are adding that name to the calling_names list
calling_names = [name for name in names if "abhi" in name]
print(calling_names)
```

```python
calling_names = [name for name in names if len(name) > 5]
print(calling_names)
```

```python
# set comprehension is similar to list comprehension
# The main difference is that it returns a set instead of a list, 
# which means that it will automatically remove duplicate items and does not maintain the order of elements.
names = [
    "abhilov",
    "aditya",
    "jayesh",
    "aakash",
    "abhilov",
    "aditya",
    "jayesh",
    "aakash"
]

unique_names = {name for name in names if len(name) > 5}
print(unique_names)
```
